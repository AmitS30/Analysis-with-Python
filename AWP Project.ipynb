# Insights
#As we early discussed we have a problem in the dataset that all the columns having object Dtype. Columns such as price and bedroom count columns having numeric data also having the object dtype so first we have to change the dtype of the those columns and then we can perform the various operations on the dataset.

```python
df.replace("Unknown",0,inplace=True)
```

```python
df["Level"].unique()
```

```python
df['Level']=df['Level'].str.replace("+","")
```

```python
df['Level']=df['Level'].str.replace("Ground","1")
```

```python
df['Bedrooms']=df['Bedrooms'].str.replace("+","")
```

```python
df['Bathrooms']=df['Bathrooms'].str.replace("+","")
```

```python
#Change the datatypes of the columns.
```

```python
df['Price']=df['Price'].astype("float")
```

```python
df["Bedrooms"]=df["Bedrooms"].astype("float")
```

```python
df["Bathrooms"]=df["Bathrooms"].astype("float")
```

```python
df["Area"]=df["Area"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop the null values

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```
